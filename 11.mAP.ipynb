```python
import cv2
import numpy as np
import os

model = cv2.dnn.readNet('/Users/amariofausta/Documents/Kuliah/Semester 8/Bobot/yolov3custom_last.weights copy 5','/Users/amariofausta/Documents/Kuliah/Semester 8/Bobot/yolov3custom.cfg')
with open("/Users/amariofausta/Documents/Kuliah/Semester 8/Code/Dataset/dataset/classes.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
    

# tentukan path ke folder testing
folder_testing = "/Users/amariofausta/Documents/Kuliah/Semester 8/Code/test"

# inisialisasi daftar ground-truth dan prediksi
gt_list = []
pred_list = []

# loop melalui setiap file dalam folder testing
for file_name in os.listdir(folder_testing):
    print(file_name)
    # baca gambar dan inisialisasi daftar ground-truth
    img = cv2.imread(os.path.join(folder_testing, file_name))
    img = cv2.imread(os.path.join(folder_testing, file_name))
    if img is None:
        print(f"Failed to read image {file_name}")
        continue  # lanjut ke file selanjutnya jika gambar tidak berhasil dibaca

    gt_labels = []

    # baca file txt dengan ground-truth
    with open(os.path.join(folder_testing, os.path.splitext(file_name)[0] + ".txt"), "r") as f:
        for line in f:
            # parsing setiap baris untuk mendapatkan koordinat bounding box dan kelas
            class_id, x, y, w, h = map(float, line.split())
            label = classes[int(class_id)]

            # tambahkan bounding box ke daftar ground-truth
            x1 = int((x - w/2) * img.shape[1])
            y1 = int((y - h/2) * img.shape[0])
            x2 = int((x + w/2) * img.shape[1])
            y2 = int((y + h/2) * img.shape[0])
            gt_labels.append((label, x1, y1, x2, y2))

    # jalankan deteksi menggunakan model YOLOv3 pada gambar
    blob = cv2.dnn.blobFromImage(img, 1/255, (416, 416), swapRB=True, crop=False)
    model.setInput(blob)
    outputs = model.forward(model.getUnconnectedOutLayersNames())

    # inisialisasi daftar prediksi
    pred_labels = []

    # loop melalui setiap deteksi yang dihasilkan oleh model YOLOv3
    for output in outputs:
        for detection in output:
            # parsing deteksi untuk mendapatkan koordinat bounding box dan confidence score
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                label = classes[class_id]

                # tambahkan bounding box ke daftar prediksi
                x1, y1, x2, y2 = detection[:4] * np.array([img.shape[1], img.shape[0], img.shape[1], img.shape[0]])
                x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
                pred_labels.append((label, confidence, x1, y1, x2, y2))

    # tambahkan daftar ground-truth dan prediksi untuk file ini ke daftar keseluruhan
    gt_list.append(gt_labels)
    pred_list.append(pred_labels)
```
