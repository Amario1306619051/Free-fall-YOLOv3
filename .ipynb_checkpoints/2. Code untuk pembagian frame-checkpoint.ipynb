```python
import cv2
```

```python
# memuat video
cap = cv2.VideoCapture("/Users/amariofausta/Downloads/VID_20230220_111022_HSR_2407262904 (1).mp4")

FrameList = []
# memastikan video dapat dibuka
if not cap.isOpened():
    print("Error opening video file")

# membaca setiap frame dari video
while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # menampilkan frame
    cv2.imshow("Frame", frame)
    FrameList.append(frame)
    
    # menunggu tombol 'q' ditekan untuk berhenti
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# menutup jendela dan melease sumber daya
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# sekumpulan gambar
images = FrameList

# loop melalui setiap gambar
for i, img in enumerate(images):
    # menyimpan gambar
    cv2.imwrite("Frame Gambar/gambar_{}.jpg".format(i), img)
```
